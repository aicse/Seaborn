```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df =pd.read_csv("C://Users//Acorz//Desktop//simpli/Data science doc//data science update//OSL Datasets//Lesson_03//ADANIPORTS.csv",parse_dates=True)
```

```python
df.head()
```

```python
df['H-L'] = df.High - df.Low
```

```python
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
sns.set_theme()
```

```python
sns.set_style("darkgrid")
```

```python
#plotting 3D Graph
ax=plt.axes(projection='3d')
ax.scatter(df.index,df['H-L'],df['100MA'])
ax.set_xlabel('indes')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```
